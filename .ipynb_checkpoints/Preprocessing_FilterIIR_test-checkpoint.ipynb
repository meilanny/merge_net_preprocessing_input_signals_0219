```python
N = 125
temp_moving_avg = np.convolve(sig_X, np.ones(N)/N, mode='same')
sig_after_reduce = sig_X - temp_moving_avg
```

```python
sig_after_reduce[100:200]
```

```python
eps = 0.7332950830459595
clip_lo = -93.8617706298828125
clip_hi = 93.1284790039062500

sig_after_reduce_new = np.zeros(sig_after_reduce.shape)

for i in range(len(sig_after_reduce)):
    if sig_after_reduce[i] > clip_hi + 0.0000001:
        sig_after_reduce_new[i] = clip_hi + 0.0000001
    elif sig_after_reduce[i] < clip_lo:
        sig_after_reduce_new[i] = clip_lo
    else:
        sig_after_reduce_new[i] = sig_after_reduce[i]
        
sig_after_reduce_new = ((sig_after_reduce_new - clip_lo) / eps).astype(int)
sig_after_reduce_new = ((sig_after_reduce_new.astype(float) * eps + clip_lo) / eps).astype(int)
```
